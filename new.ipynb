```python
#to take out the borders above and below (CROP THE PHOTO),

photo=imageio.imread('C:/Users/HP/Downloads/nitish.jpg')


photo=photo[560:1230,:,:]
photo4=photo
plt.figure(figsize=(10,10))
plt.imshow(photo3)
```

```python
#to give a vignette filter,

darker=photo<120
photo[darker]=0
plt.figure(figsize=(10,10))
plt.imshow(photo)
```

```python



#to make it brighter, RGB values can be increased
bright=photo>160

photo[bright]=210
plt.figure(figsize=(10,10))
plt.imshow(photo)
```

```python
#to take out the borders above and below (CROP THE PHOTO),

photo=imageio.imread('C:/Users/HP/Downloads/nitish.jpg')


photo=photo[560:1230,:,:]

row,col,lay=photo.shape
print(row,col)

X,Y=np.ogrid[:row,:col]
centre_row=row/2
centre_col=col/2
dist_from_center = (X - centre_row)**2 + (Y - centre_col)**2
radius = (row / 2)**2
circular_mask = (dist_from_center > radius)

photo[circular_mask]=0
plt.figure(figsize=(10,10))
plt.imshow(photo[:,200:870])
```
